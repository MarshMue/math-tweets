```julia
using LinearAlgebra
using Plots
```

```julia
A = rand(10,10)
```

```julia
x = 1:10; y = rand(10); # These are the plotting data
p11 = plot(legend=false,grid=false,foreground_color_subplot=:white)
p21= plot(y, x, xflip=true, grid=false, showaxis=false, ticks=false)
p12 = plot(x, y, grid=false, showaxis=false, ticks=false)
p22 = plot(Gray.(A), ticks=false, showaxis=false)
plot(p11, p12, p21, p22, layout = (2, 2), legend = false)
```
